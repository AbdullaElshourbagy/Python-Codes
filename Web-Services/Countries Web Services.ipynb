```python
import requests
import csv
```

```python
response = requests.get('https://restcountries.eu/rest/v2/all')
data = response.json()
data
```

```python
import csv 

with open('countries.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'alpha3Code', 'area', 'region', 'population', 'main_currency', 'main_language'])
    writer.writeheader()
    for country in data:
        name = country['name']
        alpha3Code = country['alpha3Code']
        area = country['area']
        region = country['region']
        population = country['population']
        main_currency = country['currencies'][0]['name'] if len(country['currencies']) > 0 else 'none'
        main_language = country['languages'][0]['name'] if len(country['languages']) > 0 else 'none'
        writer.writerow({'name': name, 'area': area, 'alpha3Code': alpha3Code, 'region': region, 'population': population, 'main_currency': main_currency, 'main_language': main_language})
```
