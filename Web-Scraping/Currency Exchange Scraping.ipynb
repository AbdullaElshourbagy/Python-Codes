<h3> Scrap the USD To EGP Exchange rate from this website</h3> <br> <a href="https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html">Exchange Site</a> 
<br> and then use it to make a software that takes amount of USD Dollars from the user and calculate how much will it cost in EGP.



```python
import requests
from bs4 import BeautifulSoup
```

```python
dollars = float(input('Enter dollars: '))
```

```python
url='https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html'
response = requests.request('GET',url)
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
exchange_rate = float(soup.find('span', attrs={'id': 'shd2b;'}).get_text())
exchange_rate
```

```python
result = exchange_rate * dollars

print('Exchange Result >>>>')
print(str(dollars) + ' USD = ' + str(result) + ' EGP')
```
