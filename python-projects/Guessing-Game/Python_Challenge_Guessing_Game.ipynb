```python
# Type your Final Code here
from random import randint
Actual = randint(1,100)
print("---------------------------  Guessing Game---------------------------------------")
print("Actual Number is {}".format(Actual))
print("I am Guessing a number between 1 and 100")

guesses= [] #to store number


while(True):
    Guess= int(input("Enter your number : "))

    if  Guess >= 1 and Guess <=100 :
        # append the correct bound value (in bound only)
        guesses.append(Guess)
        print('IN BOUNDS')
        
        if Guess == Actual :
            print("Wow, you've guessed Correctly this time. \nYou have guessed it in {} times".format(len(guesses)))
            break  
        #get the diff to decide if far or closer to target
        Guess_Diff=abs(Guess-Actual)    
        
        #rule between 1 and 4 are closer else far
        if Guess_Diff < 5 and Guess_Diff > 0 :
            print(' you are closer to the target ')
        else:
            print(' you are far from the target ')
    else:
        print('OUT OF BOUNDS')
        continue
        
print("\n\n\n\n")
print("----------------correct value --------------------")
print("Guesses {} with length : {}".format(guesses,len(guesses)))
```
